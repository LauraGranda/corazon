## 📚 Import libraries

```python
import warnings
from pathlib import Path

import pandas as pd

warnings.simplefilter(action="ignore", category=FutureWarning)
pd.set_option("display.float_format", "{:.2f}".format)
```

## 💾 Load data

```python
INPUT_PATH: Path = Path("/home/lauragranda01/corazon/data/03_primary/corazon_explored.parquet")
df: pd.DataFrame = pd.read_parquet(INPUT_PATH, engine="pyarrow")
print(f"Dataset shape: {df.shape}")
df.head()
```

## 📊 Target (disease) vs Numerical Variables

### disease vs age

```python
variable = "age"
df.plot(
    kind="box",
    column=variable,
    by="disease",
    grid=False,
    title=f"disease vs {variable} boxplot",
);
```

```python
df[df["disease"]][variable].plot(
    kind="kde",
    label="Disease",
    legend=True,
    title=f"disease vs {variable} density plot",
)
df[~df["disease"]][variable].plot(
    kind="kde",
    label="No Disease",
    legend=True,
    xlabel=variable,
    ylabel="Density",
);
```

## 📊 Analysis of Results and Conclusions

### Strongest Predictors of Heart Disease (bivariate analysis)

**Strong discriminators** (clear separation between disease/no-disease):
1. **max_hr**: substantially lower in disease group
2. **old_peak**: substantially higher in disease group
3. **chest_pain**: asymptomatic type shows high disease rate
4. **thal**: reversable defect strongly associated with disease
5. **ca**: clear dose-response relationship with disease rate
6. **exang**: exercise-induced angina is strong indicator
7. **sex**: males have higher disease rate
8. **slope**: flat slope (2) associated with disease

**Weak discriminators**:
- **age**: modest differences
- **rest_bp**: minimal separation
- **chol**: very weak discriminator
- **fbs**: nearly identical disease rates across groups
- **rest_ecg**: modest differences

### Key Numerical Interactions
- age and max_hr show negative correlation; disease clusters at older age + lower max_hr
- old_peak and max_hr are complementary separators of disease classes
