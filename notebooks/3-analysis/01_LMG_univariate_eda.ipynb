## 📊 Analysis of Results and Conclusions

### Numerical Variables
- **age**: Normal distribution centered around 55 years, range 29-77
- **rest_bp**: Normal distribution around 130 mm Hg with some high outliers
- **chol**: Right-skewed distribution, mean ~245 mg/dl
- **max_hr**: Slight left-skew, centered around 150 bpm
- **old_peak**: Highly right-skewed, most values near 0-1

### Binary Variables
- **fbs**: Severe imbalance (88% normal, 12% elevated)
- **exang**: Moderate imbalance (71% no, 29% yes)
- **disease**: Reasonably balanced target (~44-56%)

### Categorical Variables
- **sex**: Male-biased (68:32)
- **chest_pain**: Balanced across 4 types
- **thal**: Normal dominant (56%)
- **rest_ecg**: Normal dominant (61%)
- **slope**: Type 1 dominant (52%)
- **ca**: Zero vessels dominant (65%)
