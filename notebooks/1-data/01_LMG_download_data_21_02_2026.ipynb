## 📚 Import libraries

```python
from pathlib import Path

import pandas as pd
```

## 💾 Load data

```python
def load_data(file_path: Path) -> pd.DataFrame:
    """
    Load heart disease dataset from CSV file.

    Args:
        file_path: Path object pointing to the CSV file.

    Returns:
        DataFrame with the loaded data.
    """
    return pd.read_csv(file_path)
```

```python
DATA_PATH = Path("/home/lauragranda01/corazon/data/01_raw/corazon.csv")
df = load_data(DATA_PATH)

print(f"Dataset shape: {df.shape}")
print(f"Dataset loaded successfully from: {DATA_PATH}")
```

## 👷 Data Exploration

```python
def explore_data(df: pd.DataFrame) -> None:
    """
    Display dataset information: types, sample rows, and basic statistics.

    Args:
        df: DataFrame to explore.
    """
    print("\n" + "=" * 80)
    print("DATA TYPES AND MISSING VALUES")
    print("=" * 80)
    df.info()

    print("\n" + "=" * 80)
    print("SAMPLE (10 RANDOM ROWS)")
    print("=" * 80)
    print(df.sample(10))

    print("\n" + "=" * 80)
    print("DESCRIPTIVE STATISTICS")
    print("=" * 80)
    print(df.describe())
```

```python
explore_data(df)
```

## 📊 Data Info

### Dataset Overview

The dataset contains medical examination data for patients with the following features:

- **age**: Age in years
- **sex**: Male or Female
- **chest_pain**: Chest pain type (typical, asymptomatic, nonanginal, nontypical)
- **rest_bp**: Resting blood pressure (mm Hg)
- **chol**: Serum cholesterol (mg/dl)
- **fbs**: Fasting blood sugar > 120 mg/dl (1 = true, 0 = false)
- **rest_ecg**: Resting electrocardiographic results
- **max_hr**: Maximum heart rate achieved
- **exang**: Exercise induced angina (1 = yes, 0 = no)
- **old_peak**: ST depression induced by exercise relative to rest
- **slope**: Slope of peak exercise ST segment
- **ca**: Number of major vessels (0-3) colored by fluoroscopy
- **thal**: Thalassemia type (normal, fixed defect, reversible defect)
- **disease**: Target variable (1 = presence of disease, 0 = absence)

## 💡 Save Data to Parquet Format

```python
OUTPUT_PATH = Path("/home/lauragranda01/corazon/data/02_intermediate/corazon_raw.parquet")

# Create parent directory if it does not exist
OUTPUT_PATH.parent.mkdir(parents=True, exist_ok=True)

# Save data to parquet format
df.to_parquet(OUTPUT_PATH, engine="pyarrow", index=False)

print("=" * 80)
print("DATA EXPORTED SUCCESSFULLY")
print("=" * 80)
print(f"✓ File saved to: {OUTPUT_PATH}")
print(f"✓ Total rows: {len(df)}")
print(f"✓ Total columns: {len(df.columns)}")
print("✓ File format: Parquet (optimized columnar format)")
```
